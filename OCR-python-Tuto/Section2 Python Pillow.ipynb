## 5. Add text : 

```python
# make draw object :
draw = ImageDraw.Draw(img)
```

```python
# draw.text(pos, text, color)
draw.text((100, 50), "add a text image", (255, 255, 255))
```

```python
img
```

## 6. Add Pading to image :

```python
width, height = img.size
left, rigth = 10, 10
top, bottom = 10, 10
```

```python
new_width = width + left + rigth
new_height = height + top + bottom 
```

```python
rt_padding = Image.new(img.mode, (new_width,new_height),(0,0,255))
```

```python
rt_padding.paste(img, (left, top))
rt_padding
```

## 7. Add Blur Image :

### BoxBlur

### Gaussian Blur :

```python
img
```

### BLUR :

```python
img
```
