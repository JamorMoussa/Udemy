# Section 3: OpenCV processing text images for OCR

```python
# importing libraries :
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cv2_imchow(img, cmap=None):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR), cmap=cmap)
```

## 1. Open image with OpenCv:

```python
img = cv2.imread("./text_image.png")
```

```python
cv2_imchow(img)
```

## 2. Invert Image:

```python
inv_img = cv2.bitwise_not(img)
cv2_imchow(inv_img)
```

## 3. Banarization Image:

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
thresh, imgb = cv2.threshold(img_gray, 200, 200, cv2.THRESH_BINARY)
```

```python
cv2_imchow(imgb)
```

## 4. Erosion and Dilation:

## 4.1 Erosion :

```python
def thin_font(img):
    tmp_img = cv2.bitwise_not(img)
    kernel = np.ones((2, 2), np.uint8)
    tmp_img = cv2.erode(tmp_img, kernel, iterations=1)
    return cv2.bitwise_not(tmp_img)
```

```python
th_img = thin_font(img)
```

```python
cv2_imchow(th_img)
```

## 4.2 Delation :

```python
def thick_font(img):
    tmp_img = cv2.bitwise_not(img)
    kernel = np.ones((2,2),np.uint8)
    tmp_img = cv2.dilate(tmp_img, kernel, iterations=1)
    return cv2.bitwise_not(tmp_img)
```

```python
tck_img = thick_font(img)
```

```python
cv2_imchow(tck_img)
```
